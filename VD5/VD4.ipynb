## Sampled randomly 2000 rows out of the test data

```python
df = pd.read_csv("/content/drive/MyDrive/VD4/fashion-mnist_test.csv")
X = df.iloc[:, 1:]
y = df.iloc[:, :1]

X = X.sample(frac=0.2)
y = y.loc[X.index, :]
```

## TSNE

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(X)
```

```python
tsne_results
```

## MDS

```python
mds = MDS(n_components=2, n_init=12, max_iter=400, metric=True, n_jobs=4, random_state=2019)
X_mds = mds.fit_transform(X)
```

```python
X_mds
```

## Export data to json

```python
lists_pca = X_r.tolist()
list_tsne = tsne_results.tolist()
list_mds = X_mds.tolist()
list_y = y["label"].values.tolist()
res = {"pca": lists_pca, "tsne": list_tsne, "mds": list_mds, "y": list_y}
```

```python
with open('res.json', 'w') as fp:
    json.dump(res, fp)
```
